1.1 Initial Prompt and Pseudocode(8 marks)
My initial prompt:

I need to create a CLI Task Manager programme in Python for my programming course. The program should allow users to add,list and remove task stored in the memory. Please provide a pseudocode that only uses basic Python concepts suitable for our introductory course - not an advanced level features.

```python
# This program allows users to add, list, and remove tasks stored in memory

def main():
    print("Welcome to Task Manager!")

    # Create an empty list to store the tasks
    tasks = []

    # The program runs continuously until the user chooses to exit
    while True:
        # Display menu options for the user
        print("\nPlease choose an option:")
        print("1. Add a task")
        print("2. List all tasks")
        print("3. Remove a task")
        print("4. Exit")

        # Get user's menu choice
        choice = input("Enter your choice (1-4): ")

        # ---------------------------
        # Option 1: Add a new task
        # ---------------------------
        if choice == "1":
            task = input("Enter task description: ")

            # Check if the user entered a non-empty task
            if task.strip() == "":
                print("Task cannot be empty. Please try again.")
            else:
                tasks.append(task)
                print(" Task added successfully!")

        # ---------------------------
        # Option 2: List all tasks
        # ---------------------------
        elif choice == "2":
            if len(tasks) == 0:
                print("No tasks available.")
            else:
                print("\n Your Tasks:")
                # Display all tasks with numbers
                for index, task in enumerate(tasks, start=1):
                    print(f"{index}. {task}")

        # ---------------------------
        # Option 3: Remove a task
        # ---------------------------
        elif choice == "3":
            if len(tasks) == 0:
                print("No tasks to remove.")
            else:
                print("\nYour Tasks:")
                # Display all tasks before asking for input
                for index, task in enumerate(tasks, start=1):
                    print(f"{index}. {task}")

                # Ask user which task to remove
                task_number = input("Enter the task number to remove: ")

                # Validate that input is a number
                if task_number.isdigit():
                    task_number = int(task_number)

                    # Check if the number is within the valid range
                    if 1 <= task_number <= len(tasks):
                        removed_task = tasks.pop(task_number - 1)
                        print(f"Removed task: {removed_task}")
                    else:
                        print("Invalid task number. Please try again.")
                else:
                    print("Please enter a valid number.")

        # ---------------------------
        # Option 4: Exit the program
        # ---------------------------
        elif choice == "4":
            print("ðŸ‘‹ Goodbye! Thank you for using Task Manager.")
            break  # Exit the while loop and end the program

        # ---------------------------
        # Invalid Menu Option
        # ---------------------------
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

# Run the main function to start the program
main()
```
