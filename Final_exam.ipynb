1.1 Initial Prompt and Pseudocode(8 marks)
My initial prompt:

I need to create a CLI Task Manager programme in Python for my programming course. The program should allow users to add,list and remove task stored in the memory. Please provide a pseudocode that only uses basic Python concepts suitable for our introductory course - not an advanced level features.

```python
# This program allows users to add, list, and remove tasks stored in memory

def main():
    print("Welcome to Task Manager!")

    # Create an empty list to store the tasks
    tasks = []

    # The program runs continuously until the user chooses to exit
    while True:
        # Display menu options for the user
        print("\nPlease choose an option:")
        print("1. Add a task")
        print("2. List all tasks")
        print("3. Remove a task")
        print("4. Exit")

        # Get user's menu choice
        choice = input("Enter your choice (1-4): ")

        # ---------------------------
        # Option 1: Add a new task
        # ---------------------------
        if choice == "1":
            task = input("Enter task description: ")

            # Check if the user entered a non-empty task
            if task.strip() == "":
                print("Task cannot be empty. Please try again.")
            else:
                tasks.append(task)
                print(" Task added successfully!")

        # ---------------------------
        # Option 2: List all tasks
        # ---------------------------
        elif choice == "2":
            if len(tasks) == 0:
                print("No tasks available.")
            else:
                print("\n Your Tasks:")
                # Display all tasks with numbers
                for index, task in enumerate(tasks, start=1):
                    print(f"{index}. {task}")

        # ---------------------------
        # Option 3: Remove a task
        # ---------------------------
        elif choice == "3":
            if len(tasks) == 0:
                print("No tasks to remove.")
            else:
                print("\nYour Tasks:")
                # Display all tasks before asking for input
                for index, task in enumerate(tasks, start=1):
                    print(f"{index}. {task}")

                # Ask user which task to remove
                task_number = input("Enter the task number to remove: ")

                # Validate that input is a number
                if task_number.isdigit():
                    task_number = int(task_number)

                    # Check if the number is within the valid range
                    if 1 <= task_number <= len(tasks):
                        removed_task = tasks.pop(task_number - 1)
                        print(f"Removed task: {removed_task}")
                    else:
                        print("Invalid task number. Please try again.")
                else:
                    print("Please enter a valid number.")

        # ---------------------------
        # Option 4: Exit the program
        # ---------------------------
        elif choice == "4":
            print("Goodbye! Thank you for using Task Manager.")
            break  # Exit the while loop and end the program

        # ---------------------------
        # Invalid Menu Option
        # ---------------------------
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

# Run the main function to start the program
main()
```

```python

# Allows users to add, list, and remove tasks stored in memory

tasks = []  # Global list to store tasks

# Function to add a new task
def add_task():
    task = input("Enter a new task: ").strip()
    if task == "":
        print("Task cannot be empty.")
    else:
        tasks.append(task)
        print(f"Added: {task}")

# Function to remove a task by its number
def remove_task():
    if not tasks:
        print("No tasks to remove.")
        return

    list_tasks()
    try:
        index = int(input("Enter the task number to remove: ")) - 1
        if 0 <= index < len(tasks):
            removed = tasks.pop(index)
            print(f"Removed: {removed}")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Please enter a valid number.")

# Function to display all tasks
def list_tasks():
    if not tasks:
        print("No tasks available.")
    else:
        print("\nYour Tasks:")
        for i, t in enumerate(tasks, start=1):
            print(f"{i}. {t}")
        print()  # Blank line for readability

# Main menu function
def main():
    while True:
        print("\n===== TASK MANAGER =====")
        print("1. Add Task")
        print("2. List Tasks")
        print("3. Remove Task")
        print("4. Exit")

        choice = input("Enter your choice (1–4): ").strip()

        if choice == "1":
            add_task()
        elif choice == "2":
            list_tasks()
        elif choice == "3":
            remove_task()
        elif choice == "4":
            print("Goodbye! Your session has ended.")
            break
        else:
            print("Invalid choice. Please enter 1, 2, 3, or 4.")

# Run the program
if __name__ == "__main__":
    main()
```

```python
def refined_safe_weather_data_fetch(city):
    """
    Fetch weather data with basic error handling - Week 8 style

    >>> refined_safe_weather_data_fetch("")
    Error: City name cannot be empty
    >>> refined_safe_weather_data_fetch("InvalidCity123")  # doctest: +SKIP
    Error: Could not connect to weather service
    """

    # Step 1: Validate input
    if not city:
        print("Error: City name cannot be empty")
        return None

    try:
        # Step 2: Request weather data
        import requests
        url = f"http://wttr.in/{city}?format=j1"
        response = requests.get(url)
        data = response.json()

        # Step 3: Extract key information
        current = data['current_condition'][0]
        weather_info = {
            'city': city,
            'temperature': current['temp_C'],
            'wind_speed': current['windspeedKmph'],
            'description': current['weatherDesc'][0]['value']
        }
        return weather_info

    except:
        # Step 4: Handle connection or parsing errors
        print("Error: Could not connect to weather service")
        return None
```

```python
# Week 4–6 CLI Task Manager Program
# Demonstrates input/output, simple loops, and basic validation

tasks = []  # Global task list - Week 4 variable usage


def display_menu():
    """
    Display menu options - use Week 3 formatting techniques only

    >>> display_menu()  # doctest: +SKIP
    1. Add task
    2. List tasks
    3. Remove task
    4. Quit
    """
    print("1. Add task")
    print("2. List tasks")
    print("3. Remove task")
    print("4. Quit")


def get_user_choice():
    """
    Get and validate user choice - Week 6 input validation style

    >>> # Test case for valid choice (simulate manually)
    >>> # get_user_choice()  # doctest: +SKIP
    >>> # Expected: Returns 1, 2, 3, or 4 if valid input is given
    """
    while True:
        choice = input("Enter your choice (1–4): ").strip()

        # Week 6 concept: Input validation using if/else
        if choice in ["1", "2", "3", "4"]:
            return int(choice)
        else:
            print("Invalid input. Please enter a number between 1 and 4.")


def main():
    """Main programme loop - Week 5 loop patterns only"""
    print("Welcome to the Task Manager (Week 4–6 Version)")

    while True:  # Week 5 loop concept
        display_menu()
        choice = get_user_choice()

        # Week 5 conditional control flow
        if choice == 1:
            task = input("Enter new task: ").strip()
            if task:  # Week 6 validation
                tasks.append(task)
                print(f"Added: {task}")
            else:
                print("Error: Task cannot be empty.")

        elif choice == 2:
            if not tasks:
                print("No tasks available.")
            else:
                print("Your tasks:")
                for i, task in enumerate(tasks):
                    print(f"{i+1}. {task}")

        elif choice == 3:
            if not tasks:
                print("No tasks to remove.")
            else:
                print("Select task number to remove:")
                for i, task in enumerate(tasks):
                    print(f"{i+1}. {task}")
                try:
                    index = int(input("Enter task number: ")) - 1
                    if 0 <= index < len(tasks):
                        removed = tasks.pop(index)
                        print(f"Removed: {removed}")
                    else:
                        print("Invalid number. No task removed.")
                except:
                    print("Error: Please enter a valid number.")

        elif choice == 4:
            print("Goodbye!")
            break


# Run main program
if __name__ == "__main__":
    main()
```
