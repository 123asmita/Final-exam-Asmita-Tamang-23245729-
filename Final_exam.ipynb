1.1 Initial Prompt and Pseudocode(8 marks)
My initial prompt:

I need to create a CLI Task Manager programme in Python for my programming course. The program should allow users to add,list and remove task stored in the memory. Please provide a pseudocode that only uses basic Python concepts suitable for our introductory course - not an advanced level features.

```python
# This program allows users to add, list, and remove tasks stored in memory

def main():
    print("Welcome to Task Manager!")

    # Create an empty list to store the tasks
    tasks = []

    # The program runs continuously until the user chooses to exit
    while True:
        # Display menu options for the user
        print("\nPlease choose an option:")
        print("1. Add a task")
        print("2. List all tasks")
        print("3. Remove a task")
        print("4. Exit")

        # Get user's menu choice
        choice = input("Enter your choice (1-4): ")

        # ---------------------------
        # Option 1: Add a new task
        # ---------------------------
        if choice == "1":
            task = input("Enter task description: ")

            # Check if the user entered a non-empty task
            if task.strip() == "":
                print("Task cannot be empty. Please try again.")
            else:
                tasks.append(task)
                print(" Task added successfully!")

        # ---------------------------
        # Option 2: List all tasks
        # ---------------------------
        elif choice == "2":
            if len(tasks) == 0:
                print("No tasks available.")
            else:
                print("\n Your Tasks:")
                # Display all tasks with numbers
                for index, task in enumerate(tasks, start=1):
                    print(f"{index}. {task}")

        # ---------------------------
        # Option 3: Remove a task
        # ---------------------------
        elif choice == "3":
            if len(tasks) == 0:
                print("No tasks to remove.")
            else:
                print("\nYour Tasks:")
                # Display all tasks before asking for input
                for index, task in enumerate(tasks, start=1):
                    print(f"{index}. {task}")

                # Ask user which task to remove
                task_number = input("Enter the task number to remove: ")

                # Validate that input is a number
                if task_number.isdigit():
                    task_number = int(task_number)

                    # Check if the number is within the valid range
                    if 1 <= task_number <= len(tasks):
                        removed_task = tasks.pop(task_number - 1)
                        print(f"Removed task: {removed_task}")
                    else:
                        print("Invalid task number. Please try again.")
                else:
                    print("Please enter a valid number.")

        # ---------------------------
        # Option 4: Exit the program
        # ---------------------------
        elif choice == "4":
            print("Goodbye! Thank you for using Task Manager.")
            break  # Exit the while loop and end the program

        # ---------------------------
        # Invalid Menu Option
        # ---------------------------
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

# Run the main function to start the program
main()
```

```python


def load_tasks(filename):
    """Load tasks from file with error handling."""
    tasks = []
    try:
        with open(filename, "r") as file:
            for line in file:
                task = line.strip()
                if task:  # ignore empty lines
                    tasks.append(task)
        print("Tasks loaded successfully.")
    except FileNotFoundError:
        print(" No existing task file found. A new one will be created.")
    except Exception:
        print(" Error: File format incorrect or unreadable. Starting with an empty list.")
    return tasks


def save_tasks(filename, tasks):
    """Save tasks to file with error handling."""
    try:
        with open(filename, "w") as file:
            for task in tasks:
                file.write(task + "\n")
        print(" Tasks saved successfully.")
    except Exception:
        print(" Error saving tasks. Please check file permissions.")


def main():
    filename = "tasks.txt"
    tasks = load_tasks(filename)

    print("\nWelcome to Task Manager!")

    while True:
        # Display main menu
        print("\nMenu Options:")
        print("1. Add a task")
        print("2. List all tasks")
        print("3. Remove a task")
        print("4. Save and Exit")

        choice = input("Enter your choice (1-4): ")

        # Validate menu input
        if not choice.isdigit() or int(choice) not in [1, 2, 3, 4]:
            print(" Invalid choice. Please enter a number between 1 and 4.")
            continue

        choice = int(choice)


        # Option 1: Add new task

        if choice == 1:
            task = input("Enter task description: ").strip()
            if task == "":
                print(" Task cannot be empty. Please try again.")
            else:
                tasks.append(task)
                print(" Task added successfully!")


        # Option 2: List tasks

        elif choice == 2:
            if len(tasks) == 0:
                print("No tasks available.")
            else:
                print("\n Your Tasks:")
                for index, task in enumerate(tasks, start=1):
                    print(f"{index}. {task}")


        # Option 3: Remove a task

        elif choice == 3:
            if len(tasks) == 0:
                print("No tasks to remove.")
            else:
                print("\nYour Tasks:")
                for index, task in enumerate(tasks, start=1):
                    print(f"{index}. {task}")

                task_number = input("Enter the task number to remove: ")

                # Validate input

                if not task_number.isdigit():
                    print(" Please enter a valid number.")
                else:
                    task_number = int(task_number)
                    if 1 <= task_number <= len(tasks):
                        removed_task = tasks.pop(task_number - 1)
                        print(f"Removed task: {removed_task}")
                    else:
                        print("Invalid task number.")


        # Option 4: Save and Exit

        elif choice == 4:
            save_tasks(filename, tasks)
            print(" Goodbye!")
            break


# Run the program
main()
```
